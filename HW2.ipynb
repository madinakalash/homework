```python
#Задание 1 

#Для начала задаём текст
text = "тонешь тонешь не потонешь ты сломаешься однажды может выпьешь яда слижешь дважды знаешь мне уже не важно все не так уж важно"
words = text.split() #Теперь разбиваем текст на отдельные слова при помощи функции split()
bigrams = [] #Создаём пустой список
for i in range(len(words)-1): #Используем цикл for() для создания списка биграмм. Количество биграмм в тексте = количество слов в тексте - 1.
    bigrams.append([words[i],words[i+1]]) #Добавляем биграммы в список 
print(bigrams) #Распечатываем список 
```

```python
os.remove("poems.txt")
for root, dirs, files in os.walk('.'): #Проходимся по всем файлам, открываем и читаем их, записываем в новый файл
    for name in files:
        with open(os.path.join(root, name), 'r', encoding='utf-8', errors='ignore') as f: 
            if name[0] != ".": #Силой гугла я установила, что папка poems, видимо, была создана не на Mac OS. Поэтому мне в файл записывается лишняя информация. Короче говоря, мне нужно читать только те файлы, названия которых начинаются не на точку
                lyrics = f.read() #Читаем файлы 
                with open('poems.txt', 'a', encoding='utf-8', errors='ignore') as d: 
                    d.write(lyrics + '\n') #Не забываем о том, что между стихотворениями должно сохраняться расстояние
                
```

```python
#Задание 3
lyrics = """
Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!""" #1 Сохраняем текст в переменную
```

```python
print(len(lyrics)) #2 Узнаем длину текста в символах
```

```python
words = [word.strip("!?.,-") for word in lyrics.split()] #3 Очищаем текст от пунктуации и выводим список
print(words)
```

```python
print(' '.join(words)) #4 Объединяем этот список через пробел
```

```python
print([w.lower() if w.istitle() else w for w in words]) #5 Приводим все слова к нижнему регистру
```

```python
print(len(words)) #6 Узнаем длину текста в словах
```

```python
unique = set(words) #7 Узнаем количество уникальных слов.
print(len(unique))
```

```python
print(words[-10:])#8 Распечатаем последние 10 слов
```

```python
string = "\n".join(words) #9 Соединяем полученный в пункте 3 список в строку через знак переноса строки
print(string) 
```

```python
print(string[20:31]) #10 Распечатаем с 20 по 30 символ этой строки (включительно)
```

```python
poems = [(words[i], words[i+1]) for i in range(len(words)-1)] ##11 Разбиваем текст на биграммы и записываем их в переменную 
print(poems)
for w in poems:
    print(poems.count(w), "-", w) #11 Рассчитываем частотность каждой биграммы
```
