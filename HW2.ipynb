## Задание1
Необходимо выделить все биграммы в тексте и сохранить их список в какую-либо переменную. 

```python
#Для начала задаём текст
text = "тонешь тонешь не потонешь ты сломаешься однажды может выпьешь яда слижешь дважды знаешь мне уже не важно все не так уж важно"
words = text.split() #Теперь разбиваем текст на отдельные слова при помощи функции split()
bigrams = [] #Создаём пустой список
for i in range(len(words)-1): #Используем цикл for() для создания списка биграмм. Количество биграмм в тексте = количество слов в тексте - 1.
    bigrams.append([words[i],words[i+1]]) 
print(bigrams) 
```

```python
print(os.getcwd()) #Проверяем, где находимся
```

```python
os.remove("poems.txt")
for root, dirs, files in os.walk('.'): #Проходимся по всем файлам, открываем и читаем их, записываем в новый файл
    for name in files:
        with open(os.path.join(root, name), 'r', encoding='utf-8', errors='ignore') as f: 
            if name[0] != ".": #Силой гугла я установила, что папка poems, видимо, была создана не на Mac OS. Поэтому мне в файл записывается лишняя информация. Короче говоря, мне нужно читать только те файлы, названия которых начинаются не на точку
                lyrics = f.read() 
                with open('poems.txt', 'a', encoding='utf-8', errors='ignore') as d: 
                    d.write(lyrics + '\n') 
                
```

## Задание 3

Сохраняем текст в переменную (1)

```python

lyrics = """
Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!""" 
```

Узнаем длину текста в символах (2)

```python
print(len(lyrics)) 
```

Очищаем текст от пунктуации и выводим список (3)

```python
words = [word.strip("!?.,-") for word in lyrics.split()] 
print(words)
```

Объединяем этот список через пробел (4)


```python
print(' '.join(words)) 
```

Приводим все слова к нижнему регистру (5)

```python
print([w.lower() if w.istitle() else w for w in words]) 
```

```python
poems = [(words[i], words[i+1]) for i in range(len(words)-1)] 
print(poems)
for w in poems:
    print(poems.count(w), "-", w) 
```
