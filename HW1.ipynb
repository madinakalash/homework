```python
#Задание №1
a = "   _~_    "
b = "  (o o)   "
c = " /  V  \  "
d = "/(  _  )\ "
e = "  ^^ ^^   "
penguin = int(input("Введите количество пингвинов:"))
if penguin >=1 and penguin <=9:
    print(a*penguin, b*penguin, c*penguin, d*penguin, e*penguin, sep = "\n")
else:
    print("Не верное число")
        
```
